```python
def graphic(N, M, cnt, little_ticks, medium_ticks, many_ticks):


    tick_size = 1 / cnt
    ticks = [[] for i in range(cnt)]
    ticks_trade_size = [0 for i in range(cnt)]
    for i in range(len(trades)):
        pos = int(trades_x[i] // tick_size)
        ticks[pos].append(trades['trade_price'][i])
        ticks_trade_size[pos] += trades['trade_size'][i]
    #print(ticks_trade_size)

    open_tick = np.zeros(cnt)
    close_tick = [0 for i in range(cnt)]
    min_tick = np.zeros(cnt)
    max_tick = np.zeros(cnt)
    tick_time = np.arange(0, cnt, 1)

    for i in range(len(ticks)):
        open_tick[i] = ticks[i][0]
        close_tick[i] = ticks[i][len(ticks[i]) - 1]
        min_tick[i] = min(ticks[i])
        max_tick[i] = max(ticks[i])

   
    data = pd.DataFrame(np.array([tick_time, open_tick, max_tick, min_tick, close_tick, ticks_trade_size]).T,
                       columns=['Date', 'Open', 'High', 'Low', 'Close', 'Volume'])
    data['Date'] = pd.to_datetime(data['Date'])
    data = data.set_index('Date')
    data
    
    '''
    ema_first = data['Close'].ewm(span=little_ticks).mean()
    STR_NAME_first = 'Close, ' + str(little_ticks) + '-Day EMA'
    plt.plot(tick_time, ema_first, color = '#FA9900', linestyle='--', linewidth = 1, label=STR_NAME_first)
    
    if (medium_ticks):
        ema_second = data['Close'].ewm(span=medium_ticks).mean()
        STR_NAME_second = 'Close, ' + str(medium_ticks) + '-Day EMA'
        plt.plot(tick_time, ema_second, color = '#FB7300', linestyle='--', linewidth = 1.2, label=STR_NAME_second)
    
    if (many_ticks):
        ema_third = data['Close'].ewm(span=many_ticks).mean()
        STR_NAME_third = 'Close, ' + str(many_ticks) + '-Day EMA'
        plt.plot(tick_time, ema_third, color = '#F92900', linestyle='--', linewidth = 1.4, label=STR_NAME_third)
    '''
    
    mvc = ['#FABF00']
    mavar = [little_ticks]
    if (medium_ticks):
        mvc.append('#FB7300')
        mavar.append(medium_ticks)
    if (medium_ticks):
        mvc.append('#F92900')
        mavar.append(many_ticks)
        
    s  = mpf.make_mpf_style(mavcolors=mvc)
    save = dict(fname=('graphic_' + str(cnt) + '.png'), dpi=200, pad_inches=0.25)
    mpf.plot(data, type='candlestick', mav=mavar, volume=True, style='charles', savefig=save)
```

```python
def most_valuable_graphic(time):
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize = (20, 20))

    N = len(trades['server_timestamp'])
    GOTCHA = 0.01
    INDICATOR = time * GOTCHA

    start = min(trades['server_timestamp']) // time
    end = max(trades['server_timestamp']) // time
    x = np.arange(start, end + 1) * time + 0.5 * time
    y_trade_size = np.zeros(len(x))
    y_price_trade = np.zeros(len(x))
    color_trades = ['#A0A0A0' for i in range(N)]
    for i in range(N):
        y_trade_size[trades['server_timestamp'][i] // time - start] += trades['trade_size'][i]
    for i in range(N):
        pos = trades['server_timestamp'][i] // time - start
        contribution = trades['trade_size'][i] / y_trade_size[pos]
        y_price_trade[pos] += trades['trade_price'][i] * contribution
        if (y_trade_size[pos] >= INDICATOR):
            color_trades[i] = 'red'
    for i in range(len(x)):
        if (y_trade_size[i] == 0):
            y_price_trade[i] = y_price_trade[i - 1]

    #print(x // time)
    ax1.scatter(trades['server_timestamp'], trades['trade_price'], 
            color=color_trades, marker='o')
    ax1.plot(x, y_price_trade, color='black')
    ax2.plot(x, y_trade_size)
    
    ax3.plot(orders['timestamp'], orders['ask_price_0'] - orders['bid_price_0'], c='green')
    return y_price_trade
```

```python
y_price_trade = most_valuable_graphic(5 * 10 ** 3)
```
